```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
full_data= pd.read_csv("data.csv")
```

```python
full_data
```

```python
full_data.describe()
```

```python
full_data.drop(['Unnamed: 0'] , axis = True , inplace = True)

full_data.rename(columns = {'elavator': 'elevator' , 'Warehouse': 'warehouse'}, inplace = True)
full_data.head()
```

```python
full_data.head()
```

```python
print("maximum of rent= ", {max(full_data.rent)})
print("maximum of depost= ", {max(full_data.deposit)})
print("maximum of all deposit", {max(full_data.all_to_deposit)})
```

```python
print("maximum of area= ", {max(full_data.area)})
print("minimum of area", {min(full_data.area)})
```

```python
plt.figure(figsize=(12,10))
sns.barplot(x = full_data.rooms.value_counts(). index, y = full_data.rooms.value_counts().values )
plt.xticks(rotation = 50)
plt.title("number of rooms")
plt.show()
```

```python
plt.figure(figsize= (12,10))
sns.barplot(x= full_data.district.value_counts().index, y = full_data.district.value_counts().values)
plt.title("number of districts")
plt.xticks(rotation = 50)
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.barplot(x= full_data.floor.value_counts().index, y = full_data.floor.value_counts().values)
plt.xticks(rotation =90)
plt.title("number of floors")
plt.show()
```

```python
for col in ['floor','area','age','rooms','district']:
    sns.boxplot(x=full_data[col])
    plt.show()
```

```python
full_data = full_data[full_data.area<=200]
full_data = full_data[full_data.district<10]
full_data = full_data[full_data.floor<=6]
for col in ['area','district','floor']:
  sns.boxplot(x=full_data[col])
  plt.show()
```

```python
full_data.loc[full_data['floor'] == -1, 'floor'] = 0
```

```python
plt.figure(figsize=(12,8))
corr =full_data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

```python
sns.distplot(full_data['deposit'])
plt.show()

sns.distplot(full_data['rent'].apply(np.log1p))
plt.show()
sns.distplot(full_data['all_to_deposit'].apply(np.log1p))
plt.show()
```
